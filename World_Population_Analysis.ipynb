```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/whd.csv", skiprows=4)
df
```

```python
df.isnull().sum()
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case=False)],axis=1,inplace=True)
df.info()
```

```python
df_clean = df.dropna()
df_clean.info()
```

```python
df_clean['Country Name'].unique()
```

```python
df_clean['Indicator Name'].unique()
```

```python
df_clean.to_csv('whd_clean.csv', index=False)
```

```python
# Filter data for Aruba's total population
aruba_population = df_clean[df_clean['Country Name'] == 'Aruba'][['Year', 'SP.POP.TOTL']]

# Plot population growth over time
plt.figure(figsize=(10, 6))
plt.plot(aruba_population['Year'], aruba_population['SP.POP.TOTL'], marker='o')
plt.title('Population Growth in Aruba (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.show()
```

```python
df_clean.head()
```

```python
# Reshape the DataFrame from wide to long format
df_long = df_clean.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                  var_name='Year',
                  value_name='Value')

# Filter for the relevant indicator (e.g., total population)
df_population = df_long[df_long['Indicator Code'] == 'SP.POP.TOTL']

# Convert Year to integer
df_population['Year'] = df_population['Year'].astype(int)

# Filter for Aruba
aruba_population = df_population[df_population['Country Name'] == 'Aruba']

# Plot population growth
plt.figure(figsize=(10, 6))
plt.plot(aruba_population['Year'], aruba_population['Value'], marker='o')
plt.title('Population Growth in Aruba (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.show()
```

```python
df_long
```

```python
# Filter for male and female population percentages
df_gender_male = df_long[df_long['Indicator Code'] == 'SP.POP.TOTL.MA.ZS']
df_gender_female = df_long[df_long['Indicator Code'] == 'SP.POP.TOTL.FE.ZS']

# Filter for Aruba
aruba_gender_male = df_gender_male[df_gender_male['Country Name'] == 'Aruba']
aruba_gender_female = df_gender_female[df_gender_female['Country Name'] == 'Aruba']

# Convert Year to integer
aruba_gender_male['Year'] = aruba_gender_male['Year'].astype(int)
aruba_gender_female['Year'] = aruba_gender_female['Year'].astype(int)

# Plot gender distribution
plt.figure(figsize=(10, 6))
plt.plot(aruba_gender_male['Year'], aruba_gender_male['Value'], label='Male Population (%)', marker='o')
plt.plot(aruba_gender_female['Year'], aruba_gender_female['Value'], label='Female Population (%)', marker='o')
plt.title('Gender Distribution in Aruba (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Percentage of Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter for age dependency ratios
df_dependency_young = df_long[df_long['Indicator Code'] == 'SP.POP.DPND.YG']
df_dependency_old = df_long[df_long['Indicator Code'] == 'SP.POP.DPND.OL']

# Filter for Aruba
aruba_dependency_young = df_dependency_young[df_dependency_young['Country Name'] == 'Aruba']
aruba_dependency_old = df_dependency_old[df_dependency_old['Country Name'] == 'Aruba']

# Convert Year to integer
aruba_dependency_young['Year'] = aruba_dependency_young['Year'].astype(int)
aruba_dependency_old['Year'] = aruba_dependency_old['Year'].astype(int)

# Plot dependency ratios
plt.figure(figsize=(10, 6))
plt.plot(aruba_dependency_young['Year'], aruba_dependency_young['Value'], label='Young Dependency Ratio', marker='o')
plt.plot(aruba_dependency_old['Year'], aruba_dependency_old['Value'], label='Old Dependency Ratio', marker='o')
plt.title('Age Dependency Ratios in Aruba (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Dependency Ratio (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter for population aged 65 and above
df_aging = df_long[df_long['Indicator Code'] == 'SP.POP.65UP.TO.ZS']

# Filter for Africa Eastern and Southern
africa_aging = df_aging[df_aging['Country Name'] == 'Africa Eastern and Southern']

# Convert Year to integer
africa_aging['Year'] = africa_aging['Year'].astype(int)

# Plot aging population trend
plt.figure(figsize=(10, 6))
plt.plot(africa_aging['Year'], africa_aging['Value'], marker='o')
plt.title('Aging Population in Africa Eastern and Southern (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population Aged 65+ (%)')
plt.grid(True)
plt.show()
```

```python
# Filter for net migration
df_migration = df_long[df_long['Indicator Code'] == 'SM.POP.NETM']

# Filter for Africa Eastern and Southern
africa_migration = df_migration[df_migration['Country Name'] == 'Africa Eastern and Southern']

# Convert Year to integer
africa_migration['Year'] = africa_migration['Year'].astype(int)

# Plot net migration over time
plt.figure(figsize=(10, 6))
plt.plot(africa_migration['Year'], africa_migration['Value'], marker='o')
plt.title('Net Migration in Africa Eastern and Southern (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Net Migration')
plt.grid(True)
plt.show()
```

```python
# Filter for working-age population (15-64)
df_working_age = df_long[df_long['Indicator Code'] == 'SP.POP.1564.TO.ZS']

# Filter for Aruba
aruba_working_age = df_working_age[df_working_age['Country Name'] == 'Aruba']

# Convert Year to integer
aruba_working_age['Year'] = aruba_working_age['Year'].astype(int)

# Plot working-age population trend
plt.figure(figsize=(10, 6))
plt.plot(aruba_working_age['Year'], aruba_working_age['Value'], marker='o')
plt.title('Working-Age Population in Aruba (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Working-Age Population (%)')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Filter for age groups
df_age_male = df_long[df_long['Indicator Code'].isin(['SP.POP.0014.MA.ZS', 'SP.POP.1564.MA.ZS', 'SP.POP.65UP.MA.ZS'])]
df_age_female = df_long[df_long['Indicator Code'].isin(['SP.POP.0014.FE.ZS', 'SP.POP.1564.FE.ZS', 'SP.POP.65UP.FE.ZS'])]

# Filter for Aruba and a specific year (e.g., 2020)
year = 2020
aruba_age_male = df_age_male[(df_age_male['Country Name'] == 'Aruba') & (df_age_male['Year'] == str(year))]
aruba_age_female = df_age_female[(df_age_female['Country Name'] == 'Aruba') & (df_age_female['Year'] == str(year))]

# Extract values
male_values = [aruba_age_male[aruba_age_male['Indicator Code'] == 'SP.POP.0014.MA.ZS']['Value'].values[0],
               aruba_age_male[aruba_age_male['Indicator Code'] == 'SP.POP.1564.MA.ZS']['Value'].values[0],
               aruba_age_male[aruba_age_male['Indicator Code'] == 'SP.POP.65UP.MA.ZS']['Value'].values[0]]

female_values = [aruba_age_female[aruba_age_female['Indicator Code'] == 'SP.POP.0014.FE.ZS']['Value'].values[0],
                 aruba_age_female[aruba_age_female['Indicator Code'] == 'SP.POP.1564.FE.ZS']['Value'].values[0],
                 aruba_age_female[aruba_age_female['Indicator Code'] == 'SP.POP.65UP.FE.ZS']['Value'].values[0]]

# Create population pyramid
age_groups = ['0-14', '15-64', '65+']
y = np.arange(len(age_groups))

plt.figure(figsize=(10, 6))
plt.barh(y, male_values, color='blue', label='Male')
plt.barh(y, -np.array(female_values), color='red', label='Female')
plt.yticks(y, age_groups)
plt.title(f'Population Pyramid in Aruba ({year})')
plt.xlabel('Percentage of Population')
plt.legend()
plt.show()
```

```python
# Filter for total population
df_population = df_long[df_long['Indicator Code'] == 'SP.POP.TOTL']

# Filter for Aruba and Africa Eastern and Southern
aruba_population = df_population[df_population['Country Name'] == 'Aruba']
africa_population = df_population[df_population['Country Name'] == 'Africa Eastern and Southern']

# Convert Year to integer
aruba_population['Year'] = aruba_population['Year'].astype(int)
africa_population['Year'] = africa_population['Year'].astype(int)

# Plot comparison
plt.figure(figsize=(10, 6))
plt.plot(aruba_population['Year'], aruba_population['Value'], label='Aruba', marker='o')
plt.plot(africa_population['Year'], africa_population['Value'], label='Africa Eastern and Southern', marker='o')
plt.title('Population Growth Comparison: Aruba vs. Africa Eastern and Southern')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Prepare data for Aruba's population
aruba_population = df_population[df_population['Country Name'] == 'Aruba']
X = aruba_population[['Year']]
y = aruba_population['Value']

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict future population (e.g., 2030)
future_year = 2030
predicted_population = model.predict([[future_year]])

print(f'Predicted population for Aruba in {future_year}: {predicted_population[0]:.0f}')
```
